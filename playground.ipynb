```python
import os # file access
import magic # for file type detection
import librosa # for audio analysis
from mutagen.mp3 import MP3  
from mutagen.easyid3 import EasyID3  
import mutagen.id3  
from mutagen.id3 import ID3, TIT2, TIT3, TALB, TPE1, TRCK, TYER  
import IPython.display as ipd

MUSIC_DIR = "/home/gummyhz/Music/test"
```

## Adding ID3 tags to MP3 files

```python
if (os.path.isdir(MUSIC_DIR)):
    files = os.listdir(MUSIC_DIR)
    for file in files:
        fp = f"{MUSIC_DIR}/{file}"
        filetype = magic.from_file(fp, mime=True)
        if (filetype == "audio/mpeg"):       
            # audio/mpeg
            y, sr = librosa.load(fp)
            tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
            print(fp)
            print(tempo)
            bpm = str(int(tempo[0]))
            print(bpm)
            audio = MP3(fp, ID3=EasyID3)  
            audio['bpm'] = [bpm]
            audio['genre'] = ['hi']
            audio.save()
            audio.save(v2_version=3) 

            for tag, value in audio.items():
                print(f"{tag.capitalize()}: {', '.join(value)}")

            print('mp3 tags:', *sorted(EasyID3.Set.keys()), end='\n\n')

            ipd.Audio(fp)
```
